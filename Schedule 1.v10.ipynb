الكود

```python


# إعداد بعض الهياكل المساعدة
course_list = []
course_to_dept_level = {}
for dept_level, courses in course_department:
    course_list.extend(courses)
    for course in courses:
        course_to_dept_level[course] = dept_level

all_doctors = list(professor_course.keys())
dept_levels = [dept_level for dept_level, _ in course_department]

# إنشاء نموذج
model = cp_model.CpModel()

# عدد الأسابيع (أسبوع واحد)
num_weeks = 1
weeks = list(range(1, num_weeks + 1))

# تعريف المتغيرات الأساسية للمقررات العادية
course_schedule = {}
for course in course_list:
    for week in weeks:
        for day in days:
            for time in time_periods:
                for room in rooms:
                    var = model.NewBoolVar(f'{course}_W{week}_{day}_{time}_{room}')
                    course_schedule[(course, week, day, time, room)] = var

# متغيرات لعدد المحاضرات لكل دكتور في اليوم
doctor_day_lectures = {}
for doctor in all_doctors:
    for day in days:
        var = model.NewIntVar(0, len(time_periods), f'{doctor}_{day}_lecture_count')
        doctor_day_lectures[(doctor, day)] = var

# متغيرات لعدد المحاضرات لكل قسم في اليوم
dept_day_lectures = {}
for dept_level in dept_levels:
    for day in days:
        var = model.NewIntVar(0, len(time_periods) * len(weeks), f'{dept_level}_{day}_lecture_count')
        dept_day_lectures[(dept_level, day)] = var

# القيود الأساسية
# تقييد المقررات النظرية للقاعات النظرية والعملية للقاعات العملية
for course in course_list:
    course_category = course_type[course]  # نوع المقرر
    for week in weeks:
        for day in days:
            for time in time_periods:
                for room in rooms:
                    room_category = room_type[room]  # نوع القاعة
                    if course_category != room_category:
                        # إذا كان نوع القاعة لا يطابق نوع المقرر، نحظر هذا الجدول
                        model.Add(course_schedule[(course, week, day, time, room)] == 0)
# 1. تخصيص كل مقرر مرة واحدة في الأسبوع
for course in course_list:
    model.Add(
        sum(
            course_schedule[(course, week, day, time, room)]
            for week in weeks
            for day in days
            for time in time_periods
            for room in rooms
        ) == 1
    )

# 2. عدم تداخل المقررات في نفس القاعة والفترة
for week in weeks:
    for day in days:
        for time in time_periods:
            for room in rooms:
                model.Add(
                    sum(
                        course_schedule[(course, week, day, time, room)]
                        for course in course_list
                    ) <= 1
                )

# 3. سعة القاعة واستيعابها لعدد الطلاب
for course in course_list:
    dept_level = course_to_dept_level[course]
    num_students = number_of_student[dept_level]
    for week in weeks:
        for day in days:
            for time in time_periods:
                for room in rooms:
                    if num_students > room_capacity[room]:
                        model.Add(course_schedule[(course, week, day, time, room)] == 0)

# 4. توافر الدكتور وجدولة المقررات
for doctor, courses in professor_course.items():
    availability = doctor_availability[doctor]
    for course in courses:
        for week in weeks:
            valid_times = []
            for day in days:
                for idx, t in enumerate(time_periods):
                    if availability.get(day, [0]*len(time_periods))[idx]:
                        for room in rooms:
                            var = course_schedule[(course, week, day, t, room)]
                            valid_times.append(var)
                    else:
                        for room in rooms:
                            model.Add(course_schedule[(course, week, day, t, room)] == 0)
            # يجب أن يتم جدولة المقرر مرة واحدة
            model.Add(sum(valid_times) == 1)

# 5. عدم تداخل المحاضرات للأستاذ الواحد في نفس الوقت
for doctor, courses in professor_course.items():
    for week in weeks:
        for day in days:
            for time in time_periods:
                vars = []
                for course in courses:
                    for room in rooms:
                        vars.append(course_schedule[(course, week, day, time, room)])
                model.Add(sum(vars) <= 1)

# 6. عدم تداخل المحاضرات لطلاب المستوى الواحد في نفس الوقت
for dept_level, courses in course_department:
    for week in weeks:
        for day in days:
            for time in time_periods:
                vars = []
                for course in courses:
                    for room in rooms:
                        vars.append(course_schedule[(course, week, day, time, room)])
                model.Add(sum(vars) <= 1)

# 7. لا يمكن للدكتور تدريس أكثر من ثلاث محاضرات في اليوم الواحد
for doctor in all_doctors:
    for day in days:
        lectures = []
        for c in professor_course[doctor]:
            for week in weeks:
                for time in time_periods:
                    for room in rooms:
                        lectures.append(course_schedule[(c, week, day, time, room)])
        model.Add(doctor_day_lectures[(doctor, day)] == sum(lectures))
        model.Add(doctor_day_lectures[(doctor, day)] <= 3)

# 8. يجب أن يكون لكل قسم محاضرتان على الأقل وثلاث محاضرات كحد أقصى في اليوم إذا كان لديه محاضرات
for dept_level in dept_levels:
    for day in days:
        lectures = []
        dept_courses = [c for c in course_list if course_to_dept_level[c] == dept_level]
        for course in dept_courses:
            for week in weeks:
                for time in time_periods:
                    for room in rooms:
                        lectures.append(course_schedule[(course, week, day, time, room)])
        model.Add(dept_day_lectures[(dept_level, day)] == sum(lectures))
        # متغير يدل على أن هناك محاضرات في هذا اليوم للقسم
        is_day_has_lectures = model.NewBoolVar(f'{dept_level}_{day}_has_lectures')
        model.Add(dept_day_lectures[(dept_level, day)] >= 1).OnlyEnforceIf(is_day_has_lectures)
        model.Add(dept_day_lectures[(dept_level, day)] == 0).OnlyEnforceIf(is_day_has_lectures.Not())
        model.Add(dept_day_lectures[(dept_level, day)] >= 2).OnlyEnforceIf(is_day_has_lectures)
        model.Add(dept_day_lectures[(dept_level, day)] <= 3).OnlyEnforceIf(is_day_has_lectures)

# ====================================================================
# إضافة القيد الجديد: جدولة المقررات المشتركة بين الأقسام
# --------------------------------------------------------------------
combined_course_schedule = {}
splitted_course_schedule = {}
sc_is_combined_vars = {}  # لتخزين المتغيرات sc_is_combined لكل مقرر مشترك

for sc, info in shared_courses.items():
    departments = info['departments']
    sc_doctor = info['doctor']
    total_students = sum(number_of_student[dept] for dept in departments)
    suitable_rooms = [r for r in rooms if room_capacity[r] >= total_students]

    # متغير يحدد هل المقرر المشترك سيتم بشكل مشترك (combined) أو منفصل (splitted)
    sc_is_combined = model.NewBoolVar(f'{sc}_is_combined')
    sc_is_combined_vars[sc] = sc_is_combined

    # إذا كان هناك قاعات مناسبة تسع الجميع، نعطي أولوية مطلقة للمحاضرة المشتركة
    if suitable_rooms:
        model.Add(sc_is_combined == 1)
    else:
        model.Add(sc_is_combined == 0)

    # تعريف المتغيرات للمحاضرة المشتركة
    for week in weeks:
        for day in days:
            for time in time_periods:
                for room in rooms:
                    var = model.NewBoolVar(f'{sc}_combined_W{week}_{day}_{time}_{room}')
                    combined_course_schedule[(sc, week, day, time, room)] = var

    # تعريف المتغيرات للمحاضرات المنفصلة
    for dept_level in departments:
        for week in weeks:
            for day in days:
                for time in time_periods:
                    for room in rooms:
                        var = model.NewBoolVar(f'{sc}_splitted_{dept_level}_W{week}_{day}_{time}_{room}')
                        splitted_course_schedule[(sc, dept_level, week, day, time, room)] = var

    # قيد المحاضرة المشتركة: إذا sc_is_combined=1، يجب جدولة محاضرة مشتركة واحدة بالضبط في قاعة مناسبة
    if suitable_rooms:
        model.Add(
            sum(
                combined_course_schedule[(sc, week, day, time, room)]
                for week in weeks
                for day in days
                for time in time_periods
                for room in suitable_rooms
            ) == sc_is_combined
        )
    else:
        # لا توجد قاعة مناسبة للمحاضرة المشتركة
        model.Add(
            sum(
                combined_course_schedule[(sc, week, day, time, room)]
                for week in weeks
                for day in days
                for time in time_periods
                for room in rooms
            ) == 0
        )

    # إذا sc_is_combined=0، يجب لكل قسم محاضرة واحدة لهذا المقرر
    for dept_level in departments:
        model.Add(
            sum(
                splitted_course_schedule[(sc, dept_level, week, day, time, room)]
                for week in weeks
                for day in days
                for time in time_periods
                for room in rooms
            ) == (1 - sc_is_combined)
        )

    # توافر الدكتور
    availability = doctor_availability[sc_doctor]
    for week in weeks:
        for day in days:
            for time_idx, time in enumerate(time_periods):
                # إذا الدكتور غير متاح في هذه الفترة:
                if availability.get(day, [0]*len(time_periods))[time_idx] == 0:
                    # لا يمكن للمحاضرة المشتركة أن تكون في هذه الفترة
                    for room in rooms:
                        model.Add(combined_course_schedule[(sc, week, day, time, room)] == 0)
                    # لا يمكن أيضاً للمحاضرات المنفصلة أن تكون في هذه الفترة
                    for dept_level in departments:
                        for room in rooms:
                            model.Add(splitted_course_schedule[(sc, dept_level, week, day, time, room)] == 0)

    # عدم تداخل المحاضرات للدكتور في نفس الوقت
    for week in weeks:
        for day in days:
            for time in time_periods:
                combined_vars = [combined_course_schedule[(sc, week, day, time, r)] for r in rooms]
                splitted_vars = []
                for dept_level in departments:
                    for r in rooms:
                        splitted_vars.append(splitted_course_schedule[(sc, dept_level, week, day, time, r)])
                doctor_other_vars = []
                for c in professor_course[sc_doctor]:
                    for r in rooms:
                        doctor_other_vars.append(course_schedule[(c, week, day, time, r)])
                model.Add(sum(combined_vars) + sum(splitted_vars) + sum(doctor_other_vars) <= 1)

    # عدم تداخل المحاضرات للأقسام المشتركة في نفس الوقت
    for dept_level in departments:
        # جميع المقررات الخاصة بهذا القسم
        dept_courses = [c for c, dlevel in course_to_dept_level.items() if dlevel == dept_level]
        for week in weeks:
            for day in days:
                for time in time_periods:
                    dept_vars_normal = []
                    for c in dept_courses:
                        for r in rooms:
                            dept_vars_normal.append(course_schedule[(c, week, day, time, r)])
                    combined_vars = [combined_course_schedule[(sc, week, day, time, r)] for r in rooms]
                    splitted_vars = [splitted_course_schedule[(sc, dept_level, week, day, time, r)] for r in rooms]

                    model.Add(sum(dept_vars_normal) + sum(combined_vars) + sum(splitted_vars) <= 1)

    # سعة القاعة للمحاضرات المنفصلة
    for dept_level in departments:
        dept_students = number_of_student[dept_level]
        for week in weeks:
            for day in days:
                for time in time_periods:
                    for room in rooms:
                        if dept_students > room_capacity[room]:
                            model.Add(splitted_course_schedule[(sc, dept_level, week, day, time, room)] == 0)

    # سعة القاعة للمحاضرة المشتركة (إذا كانت ممكنة)
    if suitable_rooms:
        for week in weeks:
            for day in days:
                for time in time_periods:
                    for room in rooms:
                        if room not in suitable_rooms:
                            model.Add(combined_course_schedule[(sc, week, day, time, room)] == 0)


# ====================================================================

# حل النموذج
solver = cp_model.CpSolver()
solver.parameters.max_time_in_seconds = 300  # تحديد الحد الأقصى لوقت الحل بـ 5 دقائق
solver.parameters.num_search_workers = 8  # إذا كان لديك معالجات متعددة

status = solver.Solve(model)
```

عرض النتائج

```python
if status == cp_model.FEASIBLE or status == cp_model.OPTIMAL:
    # إنشاء جداول
    doctor_schedules = {doctor: [] for doctor in all_doctors}
    dept_level_schedules = {dept_level: [] for dept_level in dept_levels}

    # نتائج المقررات العادية
    for course in course_list:
        # الحصول على الدكتور
        doctor = None
        for d, courses in professor_course.items():
            if course in courses:
                doctor = d
                break

        dept_level = course_to_dept_level[course]

        for week in weeks:
            for day in days:
                for time in time_periods:
                    for room in rooms:
                        if solver.Value(course_schedule[(course, week, day, time, room)]):
                            # إضافة إلى جدول الدكتور
                            doctor_schedules[doctor].append({
                                'week': week,
                                'day': day,
                                'time': time,
                                'room': room,
                                'course': course
                            })
                            # إضافة إلى جدول المستوى
                            dept_level_schedules[dept_level].append({
                                'week': week,
                                'day': day,
                                'time': time,
                                'room': room,
                                'course': course,
                                'doctor': doctor
                            })

    # نتائج المقررات المشتركة
    for sc, info in shared_courses.items():
        departments = info['departments']
        sc_doctor = info['doctor']
        is_combined_val = solver.Value(sc_is_combined_vars[sc])

        if is_combined_val == 1:
            # محاضرة مشتركة
            for week in weeks:
                for day in days:
                    for time in time_periods:
                        for room in rooms:
                            if (sc, week, day, time, room) in combined_course_schedule and solver.Value(combined_course_schedule[(sc, week, day, time, room)]):
                                # إضافة إلى جدول الدكتور
                                doctor_schedules[sc_doctor].append({
                                    'week': week,
                                    'day': day,
                                    'time': time,
                                    'room': room,
                                    'course': sc + "_combined"
                                })
                                # إضافة إلى جداول الأقسام
                                for dept_level in departments:
                                    dept_level_schedules[dept_level].append({
                                        'week': week,
                                        'day': day,
                                        'time': time,
                                        'room': room,
                                        'course': sc + "_combined",
                                        'doctor': sc_doctor
                                    })
        else:
            # محاضرات منفصلة
            for dept_level in departments:
                for week in weeks:
                    for day in days:
                        for time in time_periods:
                            for room in rooms:
                                if solver.Value(splitted_course_schedule[(sc, dept_level, week, day, time, room)]):
                                    doctor_schedules[sc_doctor].append({
                                        'week': week,
                                        'day': day,
                                        'time': time,
                                        'room': room,
                                        'course': sc + f"_splitted_{dept_level}"
                                    })
                                    dept_level_schedules[dept_level].append({
                                        'week': week,
                                        'day': day,
                                        'time': time,
                                        'room': room,
                                        'course': sc + f"_splitted_{dept_level}",
                                        'doctor': sc_doctor
                                    })

    # عرض جدول كل دكتور
    for doctor, schedule in doctor_schedules.items():
        print(f"\nجدول الدكتور {doctor}:")
        for entry in sorted(schedule, key=lambda x: (x['week'], days.index(x['day']), time_periods.index(x['time']))):
            print(f"الأسبوع {entry['week']}, يوم {entry['day']}, الفترة {entry['time']}, المقرر {entry['course']}, القاعة {entry['room']}")
        # عرض عدد المحاضرات لكل يوم
        print("عدد المحاضرات لكل يوم:")
        for day in days:
            lecture_count = solver.Value(doctor_day_lectures[(doctor, day)])
            print(f"  يوم {day}: {lecture_count} محاضرة")

    # عرض جدول كل مستوى
    for dept_level, schedule in dept_level_schedules.items():
        print(f"\nجدول {dept_level}:")
        for entry in sorted(schedule, key=lambda x: (x['week'], days.index(x['day']), time_periods.index(x['time']))):
            print(f"الأسبوع {entry['week']}, يوم {entry['day']}, الفترة {entry['time']}, المقرر {entry['course']}, الدكتور {entry['doctor']}, القاعة {entry['room']}")
        # عرض عدد المحاضرات لكل يوم
        print("عدد المحاضرات لكل يوم:")
        for day in days:
            lecture_count = solver.Value(dept_day_lectures[(dept_level, day)])
            print(f"  يوم {day}: {lecture_count} محاضرة")
else:
    print("لا يوجد حل ممكن مع القيود المحددة.")
```
